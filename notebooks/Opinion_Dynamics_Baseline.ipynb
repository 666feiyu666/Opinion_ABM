```python
from pathlib import Path
import sys

import matplotlib.pyplot as plt
import pandas as pd
from IPython.display import display

cwd = Path.cwd().resolve()
PROJECT_ROOT = cwd if (cwd / 'main.py').exists() else cwd.parent
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

from config import DEFAULT_SEED
from experiments.base import load_baseline_outputs, run_baseline_experiment
from visualization import plot_final_opinion_distribution, plot_network_and_homophily, plot_time_series_summaries
```

```python
if baseline_results is not None:
    fig, _ = plot_time_series_summaries(baseline_results['history_df'])
    plt.show()

    fig, _ = plot_final_opinion_distribution(baseline_results['agents'], params=baseline_results['params'])
    plt.show()

    fig, _ = plot_network_and_homophily(baseline_results['G_updated'], baseline_results['pos'])
    plt.show()
else:
    print('Set RUN_BASELINE = True to regenerate the baseline figures in this notebook.')
```
